# taking input and constucting E function

```python
from sympy import *
print("enter the order")
order = int(input())
print("enter the equations")
system=[]
var_str=''
for i in range(1,order+1):
    var_str=var_str+'x' +str(i)+' '
var = symbols(var_str)
i=0
for i in range(0,order):
    system.append(sympify(input()))
i=0
E=0
for i in range(order):
    E = E + (system[i])**2
E = -0.5*E
```

# constructing eular equations by differ E wrt variables

```python
eular = []
i=0
print(E)
for i in range(order):
       eular.append(diff(E,var[i]))
print (eular,var)    
```
